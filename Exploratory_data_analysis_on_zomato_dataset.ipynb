```python

import zipfile

def unzip_file(zip_file_path, extract_dir):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)


zip_file_path = 'Zomatodataset.zip'
extract_dir = ''

unzip_file(zip_file_path, extract_dir)
```
